```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/gaji karyawan univ 2011-2022 Del.csv")

df['Earnings'].fillna(df['Earnings'].mean(), inplace=True)

if 'School' in df.columns:
    category_column = 'School'
else:
    category_column = 'Name'

category_avg_earnings = df.groupby(category_column)['Earnings'].mean().sort_values()

plt.figure(figsize=(30, 6))
plt.bar(category_avg_earnings.index, category_avg_earnings.values, color='cyan')
plt.xlabel('University')
plt.ylabel('Average Earnings')
plt.title(f'Average Earnings by {category_column}')
plt.xticks(rotation=90)
plt.show()
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler, StringIndexer, OneHotEncoder
from pyspark.ml.classification import LogisticRegression
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.sql.functions import col, when

spark = SparkSession.builder.appName('Classification Example').getOrCreate()

df = spark.read.csv("/content/output_jadi.csv", header=True, inferSchema=True)

df = df.withColumn('earnings_category',
                   when(col('Earnings') < 30000, 0)
                   .when(col('Earnings') < 60000, 1)
                   .otherwise(2))

train_data, test_data = df.randomSplit([0.8, 0.2], seed=42)

school_indexer = StringIndexer(inputCol='School', outputCol='school_index', handleInvalid='keep')
department_indexer = StringIndexer(inputCol='Department', outputCol='department_index', handleInvalid='keep')

school_encoder = OneHotEncoder(inputCol='school_index', outputCol='school_vec')
department_encoder = OneHotEncoder(inputCol='department_index', outputCol='department_vec')

assembler = VectorAssembler(inputCols=['Year', 'school_vec', 'department_vec'], outputCol='features')

lr = LogisticRegression(featuresCol='features', labelCol='earnings_category')

pipeline = Pipeline(stages=[school_indexer, department_indexer,
                            school_encoder, department_encoder,
                            assembler, lr])

model = pipeline.fit(train_data)

predictions = model.transform(test_data)

predictions.select('earnings_category', 'prediction').show(100)

evaluator = MulticlassClassificationEvaluator(labelCol='earnings_category', predictionCol='prediction', metricName='accuracy')
accuracy = evaluator.evaluate(predictions)
print(f'Test Data Accuracy: {accuracy}')

evaluator_f1 = MulticlassClassificationEvaluator(labelCol='earnings_category', predictionCol='prediction', metricName='f1')
f1_score = evaluator_f1.evaluate(predictions)
print(f'Test Data F1-Score: {f1_score}')
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler, StringIndexer, OneHotEncoder
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml import Pipeline

spark = SparkSession.builder.appName('Hyperparameter Tuning with Cross-Validation').getOrCreate()

df = spark.read.csv("/content/output_jadi.csv", header=True, inferSchema=True)

from pyspark.sql.functions import col, when

df = df.withColumn('earnings_category',
                   when(col('Earnings') < 30000, 0)
                   .when(col('Earnings') < 60000, 1)
                   .otherwise(2))

train_data, test_data = df.randomSplit([0.8, 0.2], seed=42)

school_indexer = StringIndexer(inputCol='School', outputCol='school_index', handleInvalid='skip')
department_indexer = StringIndexer(inputCol='Department', outputCol='department_index', handleInvalid='skip')

school_encoder = OneHotEncoder(inputCol='school_index', outputCol='school_vec')
department_encoder = OneHotEncoder(inputCol='department_index', outputCol='department_vec')

assembler = VectorAssembler(inputCols=['Year', 'school_vec', 'department_vec'], outputCol='features')

lr = LogisticRegression(featuresCol='features', labelCol='earnings_category')

pipeline = Pipeline(stages=[school_indexer, department_indexer,
                            school_encoder, department_encoder,
                            assembler, lr])

paramGrid = ParamGridBuilder() \
    .addGrid(lr.regParam, [0.01, 0.1, 0.5]) \
    .addGrid(lr.maxIter, [10, 50, 100]) \
    .build()

crossval = CrossValidator(estimator=pipeline,
                          estimatorParamMaps=paramGrid,
                          evaluator=MulticlassClassificationEvaluator(labelCol='earnings_category', metricName='accuracy'),
                          numFolds=3)

cv_model = crossval.fit(train_data)

best_model = cv_model.bestModel

print(f"Best regParam: {best_model.stages[-1].getRegParam()}")
print(f"Best maxIter: {best_model.stages[-1].getMaxIter()}")

predictions = best_model.transform(test_data)

predictions.select('earnings_category', 'prediction').show(105393)

evaluator = MulticlassClassificationEvaluator(labelCol='earnings_category', predictionCol='prediction', metricName='accuracy')
accuracy = evaluator.evaluate(predictions)
print(f'Test Accuracy: {accuracy}')

evaluator_f1 = MulticlassClassificationEvaluator(labelCol='earnings_category', predictionCol='prediction', metricName='f1')
f1_score = evaluator_f1.evaluate(predictions)
print(f'Test F1-Score: {f1_score}')
```
